# Plotting and visualization

```python
# import the required library
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib as mpl
from mpl_toolkits.mplot3d.axes3d import Axes3D
import numpy as np
```

```python
# simple plot example
x = np.linspace(-5, 2, 100)
y1 = x**3 + 5*x**2 + 10
y2 = 3*x**2 + 10 *x
y3 = 6*x + 10

fig, ax = plt.subplots()
ax.plot(x, y1, color='blue', label='y(x): $x^3 + 5x^2 + 10$' ,alpha=0.9, linewidth=2)
ax.plot(x, y2, color='red', label="y'(x): $3x^2 + 10x$")
ax.plot(x, y3, color='green', label="y'': $6x + 10$(x)")

ax.set_xlabel('x')
ax.set_ylabel('y', rotation=0)


ax.legend()
ax.set_title('Simple Plot Example')
```

```python
# customizing the graph
x = np.linspace(0, 50, 500)
y = np.sin(x) * np.exp(-x / 10)

fig, ax = plt.subplots(figsize=(10, 6))
ax.set_facecolor("#AD2D2D")
ax.plot(x, y, color='darkblue', linewidth=2, label='y(x): $sin(x) * exp(-x / 10)$')
```
